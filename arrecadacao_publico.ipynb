```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_arrecadacao = {
    'Jogo': ['Palmeiras x São Paulo', 'Palmeiras x Boca Juniors', 'Palmeiras x Corinthias', 'Palmeiras x Santos', 'Palmeiras x Flamengo', ],
    'Arrecadacao (R$ milhões)': [3.2, 3.4, 2.9, 2.2, 3.0]
}

df_arrecadacao = pd.DataFrame(dados_arrecadacao)

# Exibindo os dados
print(df_arrecadacao)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_arrecadacao['Jogo'], df_arrecadacao['Arrecadacao (R$ milhões)'], marker='o', color='b')
plt.xlabel('Jogo')
plt.ylabel('Arrecadação (R$ milhões)')
plt.title('Arrecadação do Palmeiras no Allianz Parque - 2023')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
dados_arrecadacao = {
    'Jogo': ['Palmeiras x São Paulo', 'Palmeiras x Boca Juniors', 'Palmeiras x Corinthians', 'Palmeiras x Santos', 'Palmeiras x Flamengo'],
    'Arrecadacao (R$ milhões)': [3.2, 3.4, 2.9, 2.2, 3.0],
    'Publico (mil pessoas)': [34, 39, 32, 27, 29]
}


df_arrecadacao = pd.DataFrame(dados_arrecadacao)


media_publico = df_arrecadacao['Publico (mil pessoas)'].mean()


plt.figure(figsize=(10, 6))
plt.bar(df_arrecadacao['Jogo'], df_arrecadacao['Arrecadacao (R$ milhões)'], color='b', label='Arrecadação')
plt.axhline(y=media_publico, color='r', linestyle='--', label='Média de Público')
plt.xlabel('Jogo')
plt.ylabel('Arrecadação (R$ milhões)')
plt.title('Arrecadação e Média de Público do Palmeiras no Allianz Parque - 2023')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
